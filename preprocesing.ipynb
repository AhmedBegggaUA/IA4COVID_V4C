```python
import pandas as pd
oxford = pd.read_csv('data/OxCGRT_latest.csv')
sigma =  pd.read_csv('datos_SVIR/casos_susceptibles_2020_2021.csv')
sigma.rename(columns={"Czechia": "Czech Republic", "Slovakia": "Slovak Republic"}, inplace=True)
gamma = pd.read_csv('datos_SVIR/waning_susceptibles_2020_2021.csv')
gamma.rename(columns={"Czechia": "Czech Republic", "Slovakia": "Slovak Republic"}, inplace=True)
# Lista con los países que se van a usar
paises = ['Argentina', 'Austria', 'Belgium', 'Bulgaria', 'Canada', 'Croatia',
       'Cyprus', 'Czech Republic', 'Denmark', 'Ecuador', 'Estonia', 'Finland',
       'France', 'Germany', 'Hungary', 'Ireland', 'Italy', 'Latvia',
       'Lithuania', 'Luxembourg', 'Netherlands',
       'Norway', 'Poland', 'Portugal', 'Slovak Republic', 'Slovenia', 'Spain',
       'Sweden', 'Switzerland', 'United States']
```

```python
# De oxfor se extraen las filas de los países que se van a usar
oxford = oxford[oxford['CountryName'].isin(paises)].reset_index(drop=True)
# Parseamos la fecha
oxford['Date'] = pd.to_datetime(oxford['Date'], format='%Y%m%d')
# Seleccionamos las fechas que sean menores a 2022-01-01
oxford = oxford[oxford['Date'] < '2022-01-01'].reset_index(drop=True)
# bb
```

```python
sigma_list = [] 
gamma_list = []
for pais in paises:
    sigma_list.append(list(sigma[pais]))
    gamma_list.append(list(gamma[pais]))
```

```python
import numpy as np
sigma_list = np.array(sigma_list).flatten() 
gamma_list = np.array(gamma_list).flatten()
```

```python
print("Comparación de lens", len(sigma_list), len(oxford), len(gamma_list))
```

```python
# Vamos a agregar las columnas de sigma y gamma a oxford
oxford['sigma'] = sigma_list
oxford['gamma'] = gamma_list
# Cambiamos la columna de confirmed_cases a confirmed_cases_waning, que será igual a 
oxford.ConfirmedCases.fillna(0, inplace=True)
oxford['ConfirmedCasesWaning'] = oxford['ConfirmedCases'] - oxford['sigma'] #- oxford['gamma']
```

```python
# Vamos a mostrar 3 columnas de oxford
# CountryName, Date, ConfirmedCasesWaning, sigma, gamma, ConfirmedCases
oxford[['CountryName', 'Date', 'ConfirmedCasesWaning', 'sigma', 'gamma', 'ConfirmedCases']].tail(20)
```

```python
# Guardamos el archivo
oxford['ConfirmedCasesOld'] = oxford['ConfirmedCases']
oxford['ConfirmedCases'] = oxford['ConfirmedCasesWaning']
# Dejamos la fecha como string
oxford['Date'] = oxford['Date'].dt.strftime('%Y%m%d')
# La pasamo de 
oxford.to_csv('data/OxCGRT_latest_waning_no_vac.csv', index=False)
```
