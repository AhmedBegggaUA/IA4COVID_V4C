```python
# Vamos a obtener el porcentaje de vacunas por fabricante y por dia
# Obtenemos la suma de los total_vaccinations por dia
total_vaccinations_per_day = vacunas_por_fabricante_sp.groupby('date')['total_vaccinations'].sum()
print(total_vaccinations_per_day)
# Sacamos el porcentaje de vacunas por fabricante y por dia
for dia in total_vaccinations_per_day.index:
    vacunas_por_fabricante_sp.loc[vacunas_por_fabricante_sp['date'] == dia, 'per_centage'] = vacunas_por_fabricante_sp.loc[vacunas_por_fabricante_sp['date'] == dia, 'total_vaccinations'] / total_vaccinations_per_day[dia]
```
