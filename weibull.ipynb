```python
import numpy as np
import matplotlib.pyplot as plt

def weibull(x, N, lambd, K):
    return (K/N)*(x/lambd)**(K-1)*np.exp(-(x/lambd)**K) * 100
def weibull_exp(x,lambd, K):
    return np.exp(-(x/lambd)**K) * 100
x_values = np.linspace(14, 500, 1000)
y_values = weibull_exp(x_values,  205.6 , 2.9)
# 1,997 166 2 147.1
y_values2 = weibull_exp(x_values,  166, 2) # CA
#  135 217 3.6
y_values3 = weibull_exp(x_values, 217, 3.6) # MO
#  140 191 2.7
y_values4 = weibull_exp(x_values, 191, 2.7) # SP
#  3,409 184.9 2.5
y_values5 = weibull_exp(x_values,  184.9, 2.5) # SV
# 1,005 206.2 2.9
y_values6 = weibull_exp(x_values,  206.2, 2.9) # GA
# 1,566 178.6 3 
y_values7 = weibull_exp(x_values,  178.6, 3) # JA
# 7,338 235.3 2.7 
y_values8 = weibull_exp(x_values,  235.3, 2.7) # PF
#  147 87.3 1.4
y_values9 = weibull_exp(x_values, 87.3, 1.4) # AI
#plt.figure(figsize=(10, 6))
#plt.plot(x_values, y_values,label='AZ')
#plt.plot(x_values, y_values2,label='CA')
#plt.plot(x_values, y_values3,label='MO')
#plt.plot(x_values, y_values4,label='SP')
#plt.plot(x_values, y_values5,label='SV')
#plt.plot(x_values, y_values6,label='GA')
#plt.plot(x_values, y_values7,label='JA')
#plt.plot(x_values, y_values8,label='PF')
#plt.plot(x_values, y_values9,label='AI')
#plt.legend()
#plt.xlabel('Days from first infection')
#plt.ylabel('Probability of infection')
#plt.title('Weibull distribution for complete dosis')
#plt.show()
# Now let's use Plotly
import seaborn as sns
sns.set_style("whitegrid")
plt.figure(figsize=(20, 10))
# En las primeras 14 semanas, la probabilidad de infección es del 100%
# Añadimos 14 100s al principio
sns.lineplot(x=x_values, y=y_values,label='Oxford/AstraZeneca')
sns.lineplot(x=x_values, y=y_values2,label='CanSino')
sns.lineplot(x=x_values, y=y_values3,label='Moderna Biotech')
sns.lineplot(x=x_values, y=y_values4,label='Sinopharm')
sns.lineplot(x=x_values, y=y_values5,label='Sinovac')
sns.lineplot(x=x_values, y=y_values6,label='Gamaleya')
sns.lineplot(x=x_values, y=y_values7,label='Janssen')
sns.lineplot(x=x_values, y=y_values8,label='Pfizer/BioNTech')
# cambiamos el primer valor de AI por un 100
# El segundo por un 97
sns.lineplot(x=x_values, y=y_values9,label='Active Infection')
plt.legend(loc='upper right', fontsize=20, title='Vaccine', title_fontsize=24)
plt.xlabel('Days from active infection (AI)/fully vaccination', fontsize=34)
plt.ylabel('Inmunity probability (%)', fontsize=34)
# En el plot, las xticks empiezan desde el día 14  
plt.xticks(np.arange(0, 500, 50), fontsize=24)
plt.yticks(np.arange(0, 101, 10), fontsize=24)
plt.show()
```
